# Exploratory Data Analysis

In this notebook we will start by doing exploratory data analysis, then predictive modeling, and in the final we will evaluate our model. This dataset can be found on [kaggle](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/overview). Let's start by import neccessary modules and our dataset.

```python
# essential
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("dark")

# decomposition
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

Our dataset set is composed by two dataframe object; train set is the place we train our model and test set where we make prediction to the regression problem. Now, we focus on train set only. Let's see the shape of train set.

```python
print(train.shape)
```

Train set contains 1460 data points and 81 features. But what are those features?

```python
print(train.columns)
```

Our feature's names are not quite informative by looking it alone, so in data_description.txt, there are specified meaning for each feature.

Are there any missing values on each feature in train set?

```python
train.columns[train.isnull().any()].shape
```

There are 19 features with missing values. Which feature are those?

```python
train.columns[train.isnull().any()]
```

Let see how much missing values for each of there features.

```python
missing = train.isnull().sum()
missing = missing[missing > 0]
missing = missing.sort_values()
ax = plt.axes()
sns.barplot(x=missing.index, y=missing, ax=ax)
ax.set_xticklabels(missing.index, rotation=45, ha='right');
```

In here we can see which feature with missing values, five features with a lot of missing values are PoolQC (Pool Quality), MiscFeature (Miscellaneous feature not covered in other categories), Alley (Type of alley access to property), Fence (Fence quality), and FireplaceQU (Fireplace quality),

Now, we want to know the distribution for each feature. Before that we separate which feature is quantitative and which feature is qualitative. Also we remove column Id and SalePrice because column Id works like index and column SalePrice is our target vector.

```python
quantitative = [f for f in train.columns if train.dtypes[f] != 'object']
quantitative.remove('SalePrice')
quantitative.remove('Id')
qualitative = [f for f in train.columns if train.dtypes[f] == 'object']
```

```python
len(quantitative)
```

```python
print(quantitative)
```

```python
len(qualitative)
```

```python
print(qualitative)
```

Here we have 36 quantitative features and 43 qualitative features. Here is our distribution for quantitative type features:

```python
f = pd.melt(train, value_vars=quantitative)
g = sns.FacetGrid(f, col="variable",  col_wrap=6, sharex=False, sharey=False)
g = g.map(sns.distplot, "value")
```

Here, for qualitative feature:

```python
for c in qualitative:
    train[c] = train[c].astype('category')
    if train[c].isnull().any():
        train[c] = train[c].cat.add_categories(['MISSING'])
        train[c] = train[c].fillna('MISSING')

def boxplot(x, y, **kwargs):
    sns.boxplot(x=x, y=y)
    x=plt.xticks(rotation=45)
f = pd.melt(train, id_vars=['SalePrice'], value_vars=qualitative)
g = sns.FacetGrid(f, col="variable",  col_wrap=4, sharex=False, sharey=False, height=5)
g = g.map(boxplot, "value", "SalePrice")
```
